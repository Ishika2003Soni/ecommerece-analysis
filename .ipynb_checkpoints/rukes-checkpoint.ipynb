```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
rules = pd.read_csv("associationRules.csv",index_col="index")
rules.head(2)
```

```python
cols  = rules.columns
```

```python
first_item =cols[0]
second_item = cols[1]
```

```python
second_item
```

```python
antecedents =rules[first_item].apply(extract_content)
```

```python
import re

def extract_content(s):
    match = re.search(r"frozenset\(\{?'(.+?)'?\}\)", s)
    return match.group(1) if match else s

```

```python
rules[first_item] = antecedents
rules.head()
```

```python
consequent =rules[second_item].apply(extract_content)
```

```python
rules[second_item]= consequent
```

```python
rules.head()
```

```python
rules.to_csv("associationRules.csv")
```
